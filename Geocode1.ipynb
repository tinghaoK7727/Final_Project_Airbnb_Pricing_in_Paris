```python
airbnb['geom'] = airbnb.apply(lambda row: (str(row.Lat),str(row.Lon)),axis=1)
```

```python
airbnb1 = airbnb.loc[airbnb['labels']==0]
airbnb1
```

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

geolocator = Nominatim(user_agent="application")

reverse = RateLimiter(geolocator.reverse, min_delay_seconds=1)
```

```python
airbnb2 = airbnb.loc[airbnb['labels']==1]
airbnb2_1 = airbnb2.loc[178:3500]
airbnb2_1['address'] = airbnb2_1.apply(lambda row: geolocator.reverse(row.geom).address,axis=1)
airbnb2_1.to_excel(r'C:\Users\gixi_\Downloads/airbnb2_1.xlsx')
```
