```python
import pandas as pd
import numpy as np
```

```python
airbnb = pd.read_excel(r'/Users/tinghaokao/Desktop/Ironhack courses/Final Project/Airbnb Data/airbnb.xlsx')
```

```python
airbnb11 = airbnb.loc[airbnb['labels']==11]
airbnb11
```

```python
# airbnb11

from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
geolocator = Nominatim(user_agent="application")
reverse = RateLimiter(geolocator.reverse, min_delay_seconds=1)
airbnb11['geom'] = airbnb11.apply(lambda row: (str(row.Lat),str(row.Lon)),axis=1)
airbnb11['address'] = airbnb11.apply(lambda row: geolocator.reverse(row.geom).address,axis=1)
airbnb11
```

```python
airbnb11.to_excel(r'/Users/tinghaokao/Desktop/Ironhack courses/Final Project/Airbnb Data/airbnb11.xlsx')
```
